```python
from parsing import parse_functions
import json
def extract_features(log_file):
    """
    Extracts the features from the log file.
    """
    # Parse the log file
    atom_data, bond_data, mol_data = parse_functions.parse_gaussian_log(log_file)
    return atom_data, bond_data, mol_data

records = []
qm_logs = qm_logs.dropna()

qm_logs = qm_logs[qm_logs["rxn"].isin(thermo_logs["rxn"])]

for _, row in qm_logs.iterrows():
    # Get the log file path
    rec = {"rxn": row["rxn"]}
    for tag in ("r0","r1","p0","p1"):
        atoms, bonds, mol = extract_features(row[tag])
        rec[f"{tag}_atoms"] = json.dumps(atoms)
        rec[f"{tag}_bonds"] = json.dumps(bonds)
        rec[f"{tag}_mol"]   = json.dumps(mol)
    records.append(rec)

out_df = pd.DataFrame(records)
out_df.to_csv("reaction_features.csv", index=False)

    
```
