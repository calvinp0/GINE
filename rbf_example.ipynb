```python
import torch

def rbf_expand(distances, centers, gamma):
    return torch.exp(-gamma * (distances[..., None] - centers) ** 2)

# Example
centers = torch.linspace(0, 5, 16)  # 16 RBF centers from 0 to 5 Ã…
gamma = 10.0
distances = torch.tensor([1.0, 1.5, 3.2])  # bond lengths

rbf_feats = rbf_expand(distances, centers, gamma)
```

```python
print("RBF features shape:", rbf_feats.shape)
```

```python
print("RBF features:" , rbf_feats)
```

```python
# Plotting the RBF features
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
for i in range(rbf_feats.shape[0]):
    plt.plot(centers.numpy(), rbf_feats[i].numpy(), label=f"Distance {distances[i].item()} Ã…")
plt.title("Radial Basis Function (RBF) Features")
plt.xlabel("RBF Centers")
plt.ylabel("RBF Value")
plt.legend()
plt.grid()
plt.show()
```
